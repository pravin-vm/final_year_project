```python
import pandas as pd

# Load the CSV files and add the year column
df1_2018 = pd.read_csv('2018.csv')
df1_2018 ['year'] = 2018

df2_2019 = pd.read_csv('2019.csv')
df2_2019['year'] = 2019

df3_2020 = pd.read_csv('2020.csv')
df3_2020['year'] = 2020
df4_2021 = pd.read_csv('2021.csv')
df4_2021['year'] = 2021
df5_2022 = pd.read_csv('2022.csv')
df5_2022['year'] = 2022
df6_2023 = pd.read_csv('2023.csv')
df6_2023['year'] = 2023
```

```python
import pandas as pd

# Step 1: Identify common columns across all datasets
common_columns = set(df1.columns).intersection(set(df2.columns)).intersection(set(df3.columns)).intersection(set(df4.columns)).intersection(set(df5.columns)).intersection(set(df6.columns))

# Step 2: Filter datasets to include only common columns
df1_filtered = df1[list(common_columns)]
df2_filtered = df2[list(common_columns)]
df3_filtered = df3[list(common_columns)]
df4_filtered = df4[list(common_columns)]
df5_filtered = df5[list(common_columns)]
df6_filtered = df6[list(common_columns)]

# Step 3: Concatenate or merge filtered datasets into a new DataFrame
common_df = pd.concat([df1_filtered, df2_filtered, df3_filtered, df4_filtered, df5_filtered, df6_filtered], ignore_index=True)
```

**ML Models**

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split,cross_val_score,StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import matplotlib.pyplot as plt

# Assuming common_df is already defined

# Identify categorical columns and numerical columns
categorical_columns = common_df.select_dtypes(include=['object']).columns.tolist()
numerical_columns = common_df.select_dtypes(include=['int', 'float']).columns.tolist()

# Initialize and apply LabelEncoders to categorical columns
label_encoders = {col: LabelEncoder() for col in categorical_columns}
for col in categorical_columns:
    common_df[col] = label_encoders[col].fit_transform(common_df[col].astype(str))

# Combine numerical and encoded categorical columns
X = common_df[numerical_columns + categorical_columns]

# Target variable
target = 'Does your employer provide mental health benefits as part of healthcare coverage?'
y = common_df[target]

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardize the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


```

```python
y_train
```

```python
# Cross-validation setup
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Initialize models
models = {
    'KNN': KNeighborsClassifier(n_neighbors=5),
    'SVM': SVC(kernel='linear'),
    'Logistic Regression': LogisticRegression(max_iter=1000)
}

metrics = {'Accuracy': {}, 'Precision': {}, 'Recall': {}, 'F1 Score': {}}

def calculate_metrics(model_name, y_true, y_pred):
    metrics['Accuracy'][model_name] = accuracy_score(y_true, y_pred)
    metrics['Precision'][model_name] = precision_score(y_true, y_pred, average='macro')
    metrics['Recall'][model_name] = recall_score(y_true, y_pred, average='macro')
    metrics['F1 Score'][model_name] = f1_score(y_true, y_pred, average='macro')

# Evaluate models using cross-validation
for model_name, model in models.items():
    cv_results = cross_val_score(model, X_train_scaled, y_train, cv=skf, scoring='accuracy')
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)
    calculate_metrics(model_name, y_test, y_pred)
    print(f'{model_name} - CV Accuracy: {cv_results.mean():.4f} ± {cv_results.std():.4f}')


```

```python
import xgboost as xgb
import lightgbm as lgb
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_auc_score, accuracy_score

import warnings

# Suppress all warnings
warnings.filterwarnings('ignore')
```

```python
model_dict = {
    'Random Forest':RandomForestClassifier,
    'XGBoost':XGBClassifier,
    'LightGBM':LGBMClassifier
}
```

```python
def get_models(model_names):
    """
    Retrieve models from model_dict based on the given list of model names.
    """
    model_instances = []
    for name in model_names:
        if name in model_dict:
            model_class = model_dict[name]
            model_instances.append((name, model_class()))
        else:
            print(f"Warning: Model '{name}' is not in model_dict.")
    return model_instances
```

```python
def evaluate_model(model, model_name, X_train, y_train, X_test, y_test):
    """
    Train and evaluate a machine learning model with various metrics.

    """
    # Train the model
    model.fit(X_train, y_train)

    # Make predictions
    y_test_pred = model.predict(X_test)
    y_train_pred = model.predict(X_train)


    report_train = classification_report(y_train, y_train_pred, output_dict=True)



    # Extract metrics from the classification report for the train data
    accuracy_train = report_train['accuracy']
    precision_train = report_train['weighted avg']['precision']
    recall_train = report_train['weighted avg']['recall']
    f1_train = report_train['weighted avg']['f1-score']

    # Calculate AUC scores
    #auc_train, auc_test = calculate_auc(model, X_train, y_train, X_test, y_test)
    #auc_test = calculate_auc(model, X_train, y_train, X_test, y_test)

    # Prepare the metrics dictionary
    metrics = {
        'Model': model_name,
        'Train Accuracy': accuracy_train,
        'Train Precision': precision_train,
        'Train Recall': recall_train,
        'Train F1 Score': f1_train,
        #'Train AUC': auc_train
    }

    return metrics
```

```python
def calculate_auc(model, X_train, y_train, X_test, y_test):
    """
    Calculate the AUC score for both training and test data.
    """
    if hasattr(model, 'predict_proba'):
        # Calculate AUC for test data
        y_test_proba = model.predict_proba(X_test)[:, 1]
        auc_test = roc_auc_score(y_test, y_test_proba)

        # Calculate AUC for training data
        y_train_proba = model.predict_proba(X_train)[:, 1]
        auc_train = roc_auc_score(y_train, y_train_proba)

        return auc_train, auc_test
```

```python
def evaluate_all_models(model_names, X_train, y_train, X_test, y_test):
    """
    Evaluate a list of model names and compile the metrics into a DataFrame.

    """
    model_list = get_models(model_names)
    all_metrics = []

    for model_name, model in model_list:
        print(f"Evaluating {model_name}...")
        metrics = evaluate_model(model, model_name, X_train, y_train, X_test, y_test)
        all_metrics.append(metrics)

    # Convert the metrics list to a DataFrame
    metrics_df = pd.DataFrame(all_metrics)

    # Print the metrics DataFrame
    print("\nModel Evaluation Metrics:")
    print(metrics_df)

    return metrics_df

# Define model names to evaluate
model_names = [
    'Random Forest',
    'XGBoost',
    'LightGBM'
]

```
